[Mas funciones de Numpy](https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.math.html)

Tambien podemos apoyarnos de una funcion definida por nosotros para realizar cambios en los daframe que necesitemos. De esto se puede sacar mucho provecho cuando se trabaja con bases de datos a las cuales se les tiene que hacer mucho tratamiento

```python
## Función para etiquetar salarios
def Cate_Sal(salario):
    if salario<5000:
        return '<5000'
    else:
        if salario<10000:
            return '5000-10000'
        else:
            return '10000+'
```

```python
sal['Categoria_Salario']= sal.Total_Salario.apply(Cate_Sal)
sal.columns
```

```python
sal[['Categoria_Salario','Total_Salario']]
```

### Agrupar

El comando que vamos a utilizar es **df.groupby(cols)** donde:
    
**cols:** Lista de columnas con las que se quiere agrupar

Generamos un dataframe agrupando por la variable Grupo

```python
type(sal.groupby('Grupo'))
```

```python
salarios_grupo=sal.groupby('Grupo')
salarios_grupo.head(2)
```

```python
salarios_grupo[['Total_Salario']].mean()
```

Ahora lo agrupamos por dos variables

```python
grupo2=sal.groupby(['Anio','Grupo'])

grupo2.head()
```

Agrupamos algunas columans, por dos o mas variables. Puede parecer obvio pero es necesario que la columna por la cual estemos haciendo la agrupacion este dentro de las columas que queremos extraer, de lo contrario arrojara un error.

```python
grupo3= sal[['Anio','Departamento','Salario_Normal',
             'Total_Beneficios']].groupby(['Anio','Departamento'])
```

## Comando Aggregate

Este comandos nos permite generar diferentes funciones para cada columna de un dataframe agrupado y tiene la ventaja de poder usar las funciones **lambda**. Esto se consigue a partir de un diccionario en el que se indican las columnas y funciones que deben generarse:

## Funciones Lambda

```python
## Definición de una función que suma dos variables
def summa(a,b):
    su= a+b
    return(su)
```

```python
## Definición de una función que suma dos variables con funciones lambda
sum_lambda= lambda x,y: x+y

print('Función Lambda')
print(sum_lambda(10,11))
print('Función def')
print(summa(10,11))
```
