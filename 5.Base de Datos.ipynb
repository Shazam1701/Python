```python
# Cargamos el archivo para este ejercicio
base_na= pd.read_csv(dire + "\\MuestraDropNa.csv", sep='|')
base_na
```

```python
# Generamos una base borrando las columnas (axis= ) donde al menos
# un valor sea missing (how= )
base_na.dropna(axis= 1, how= 'any')
```

```python
# Borramos las columnas (axis= ) donde todos los valores sean missing 
# (how= )
base_na.dropna(axis= 1, how= 'all')
```

```python
# Borramos las filas (axis= ) donde todos los valores sean missing (how= )
base_na.dropna(axis= 0, how= 'all')
```

```python
# Borramos las filas (axis= ) donde alguno de los valores sea missing (how= )
base_na.dropna(axis= 0, how= 'any')
```

```python
# Borramos las filas (axis= ) donde alguno de los valores sea missing (how= )
# tomando en cuenta solo los NA's de las columnas "C1" y "C2"
base_na.dropna(axis= 0, how= 'any', subset=['C1', 'C2'])
```

# Imputacion de Valores

Para esta accion recurriremos al comando **df.fillna()**

```python
base_na.head()
```

```python
# Sustituimos los valores faltantes por 0
base_na.fillna(0)
```

Tambien podemos sustituir los valores faltantes por un valor diferente en cada columna. Esto se puede lograr con ayuda de un diccionario

```python
l_na={'C1': 100, 'C2': 50, 'C3': 300}
base_na.fillna(value=l_na)
```

La imputacion es un metodo estadistico muy utilizado que puede ir desde lo m√°s sencillo como el ejemplo que a continuacion se mostrara hasta lo m√°s complejo con criterios matematicos muy rigurosos. Por aras de la facilidad aludiremos esta ocasion a un ejemplo sencillo en el que imputemos algunos parametros estadisticos a cada columna:

```python
# Imprimimos diferentes m√©tricas estad√≠sticas pora cada columna
dic_met= {'C1': base_na['C1'].mean(),
          'C2': base_na['C2'].median(),
          'C3': base_na['C3'].max()}

base_na.fillna(dic_met)
```

De nuevo, el criterio puede ser muy diverso, donde generalmente para los valores numericos se usan parametros estadiscitocs como los anteriores, el calculo de una regresion con los valores que si se tienen o su calculo por medio de un metodo como el bootstrap y en el caso de variables categoricas se puede utilizar la moda de las variables segmentadas.
